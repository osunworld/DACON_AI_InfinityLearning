```python
import pandas as pd

# Base 경로 설정 (너희 팀 폴더명에 맞게 수정)
BASE_DIR = '/content/drive/MyDrive/Dacon_FakeText/'

TRAIN_CSV_PATH = BASE_DIR + 'data/train.csv'
TEST_CSV_PATH = BASE_DIR + 'data/test.csv'

# CSV 로딩
train_df = pd.read_csv(TRAIN_CSV_PATH)
test_df = pd.read_csv(TEST_CSV_PATH)

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")

# train 데이터 컬럼 확인
print(train_df.columns)
# test 데이터 컬럼 확인
print(test_df.columns)
```

```python
# 문단 나누기: \n 기준
def split_into_paragraphs(text):
    return [p for p in str(text).split('\n') if p.strip() != '']
```

```python
train_paragraphs = []

for idx, row in train_df.iterrows():
    if idx % 20000 == 0:
        print(f"Processing row {idx}/{len(train_df)}...")

    paragraphs = split_into_paragraphs(row['full_text'])
    for p_idx, paragraph in enumerate(paragraphs):
        train_paragraphs.append({
            'title': row['title'],
            'paragraph_index': p_idx,
            'paragraph_text': paragraph,
            'generated': row['generated']  # 원래 글의 라벨 그대로 사용
        })

# 최종 DataFrame으로 변환
train_paragraph_df = pd.DataFrame(train_paragraphs)
print(train_paragraph_df.head())
print(f"Total Paragraphs: {train_paragraph_df.shape[0]}")
```

```python
train_paragraph_df.to_csv(BASE_DIR + 'data/train_paragraph.csv', index=False, encoding='utf-8')
```
